```python
import pandas as pd
```

```python
data = pd.read_csv("max_filtered_data.csv", encoding='utf-8')

data
```

```python
max_date = data['strd_date'][0]

max_date
```

```python
v_types = ['승용', '승합', '특수', '화물']

# 승용 -> passenger
passenger_data = data[data['vehi_typ_dv_val'] == v_types[0]]
# 승합 -> bus
bus_data = data[data['vehi_typ_dv_val'] == v_types[1]]
# 특수&화물 -> truck
truck_data = data[(data['vehi_typ_dv_val'] == v_types[2]) | (data['vehi_typ_dv_val'] == v_types[3])]

passenger_data
```

```python
# 원본 xml 파일에서 엣지 정보 추출
from lxml import etree

tree = etree.parse("osm.passenger.trips.xml")
root = tree.getroot()

datas = root.getchildren()
e_list = []

for d in datas:
  e_list.append(d.get("to"))

# 중복 제거
e_list1 = set(e_list)
edge_list = list(e_list1)
edge_list = list(filter(None, edge_list))

edge_list
```

```python
hour = 3600
depart_list = ['360610647', '203971518'] # 광안대교로 진입하는 톨게이트 (수영강변, 벡스코)
```

```python
# csv to xml (passenger)
import xml.etree.ElementTree as ET
import xml.dom.minidom
import random

# 네임스페이스 정의
namespaces = {
    'xsi': 'http://www.w3.org/2001/XMLSchema-instance'
}
# 루트 요소 생성 및 네임스페이스 설정
root = ET.Element('routes', {
    'xmlns:xsi': namespaces['xsi'],
    'xsi:noNamespaceSchemaLocation': 'http://sumo.dlr.de/xsd/routes_file.xsd'
})

# vType
vType = ET.SubElement(root, 'vType', {
    'id': 'veh_passenger',
    'vClass': 'passenger'
})

# trip 부분
nc = 0
test_depart = 0

# data parsing
for index, row in passenger_data.iterrows():
    num_passengers = row['vehi_num']
    if num_passengers == 0 :
      continue
    part_time =  hour / num_passengers
    depart_time = test_depart
    for num in range(num_passengers):
      list_row = list(row)
      id_num = 'veh' + str(nc)
      rn_from = random.choice(depart_list)
      rn_to = random.choice(edge_list)

      trip = ET.SubElement(root, 'trip', {
          'id': id_num,
          'type': 'veh_passenger',
          'depart': str(round(depart_time, 3)),
          'departLane': 'best',
          'from': rn_from,
          'to': rn_to
      })

      nc += 1
      depart_time += part_time
    test_depart += hour


# ElementTree 객체 생성
tree = ET.ElementTree(root)
# print(tree.SubElement)

# XML 문자열로 변환
xml_str = ET.tostring(root, encoding='utf-8')

# Sorting
dom = xml.dom.minidom.parseString(xml_str)
passenger_trips_xml = dom.toprettyxml(indent="  ")

# Save
with open('passenger_trips.xml', 'w', encoding='utf-8') as f:
    f.write(passenger_trips_xml)
```
